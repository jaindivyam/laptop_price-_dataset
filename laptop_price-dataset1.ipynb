```python
import pandas as pd 
import numpy as np 
import plotly.express as px 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)
df = pd.read_csv("laptop.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df['Company'].value_counts())
sns.histplot(data = df , y = 'Company' , color = '#eb9973')
plt.title('Distribution of Companies')
plt.show()
```

```python
print(df['TypeName'].value_counts())
sns.histplot(data = df , y = 'TypeName' , hue = 'TypeName', palette = 'flare' )
plt.title('Distribution of Types')
plt.show()
```

```python
print(df['Inches'].value_counts())
sns.histplot(data=df, y='Inches', color='#eb9973', binwidth=0.5)
plt.title('Inches Distribution')
plt.show()
```

```python
print(df['CPU_Company'].value_counts())
sns.histplot(data = df , x = 'CPU_Company', hue = 'CPU_Company' , palette='flare')
plt.title('CPU Company')
plt.show()
```

```python
print(df['CPU_Frequency (GHz)'].value_counts())
sns.histplot(data = df , x = 'CPU_Frequency (GHz)',color='#eb9973')
plt.title('CPU_Frequency (GHz)')
plt.show()
```

```python
print(df['RAM (GB)'].value_counts())
sns.histplot(data = df , x = 'RAM (GB)',color='#eb9973', binwidth=2, linewidth=2)
plt.title('RAM (GB)')
plt.xticks(range(0, int(df['RAM (GB)'].max()) + 4, 4))
plt.show()
```

```python
print(df['GPU_Company'].value_counts())
sns.histplot(data = df , x = 'GPU_Company',color='#eb9973', binwidth=2, linewidth=2)
plt.title('GPU_Company')
plt.show()
```

```python
print(df['OpSys'].value_counts())
sns.histplot(data = df , y = 'OpSys',color='#eb9973', binwidth=2, linewidth=2)
plt.title('Operating System')
plt.show()
```

```python
sns.histplot(df,x = 'Weight (kg)' ,kde=True , color='#eb9973' )
plt.title('Weight (Kg)')
plt.show()
```

```python
sns.histplot(df,x = 'Price (Euro)' ,kde=True , color='#eb9973' )
plt.title('Price (Euro)')
plt.show()
```

```python
sns.boxplot(df , x = 'Company', y = 'Price (Euro)' )
plt.xticks(rotation='vertical')
plt.title('Company Vs Price')
plt.show()
```

```python
sns.barplot(data=df, x='Inches', y='Price (Euro)', color='#A020F0')
plt.xticks(rotation='vertical')
plt.title('Inches vs Price')
```

```python
plt.figure(figsize=(16,6))
sns.barplot(data=df, x='ScreenResolution', y='Price (Euro)', color='#A020F0')
plt.xticks(rotation='vertical')
plt.title('ScreenResolution vs Price')
```

```python
sns.violinplot(data=df, x='CPU_Company', y='Price (Euro)', color='#A020F0')
plt.xticks(rotation='vertical')
plt.title('CPU company vs Price')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Memory', y='Price (Euro)', color='#A020F0')
plt.xticks(rotation='vertical')
plt.title('Memory vs Price')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='RAM (GB)', y='Price (Euro)', hue = 'GPU_Company' , palette = 'rocket')
plt.xticks(rotation='vertical')
plt.title('RAM vs Price')
```

```python
sns.lmplot(data=df, x='Weight (kg)', y='Price (Euro)')
plt.xticks(rotation='vertical')
plt.title('Weight (kg) vs Price')
plt.show()
```

```python
numerical_data = df.select_dtypes(include=['int64', 'float64'])


corr_matrix = numerical_data.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```
